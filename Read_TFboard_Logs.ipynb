```python
from datetime import datetime
from time import mktime
import math


def get_step_duration(parsed_item):
    # print(parsed_item)
    steps_per_value = parsed_item['step'].diff().mean()
    n_values = parsed_item['wall_time'].shape[0]
    n_steps = steps_per_value * n_values
    
    dt_max = datetime.fromtimestamp(mktime(gmtime(parsed_item['wall_time'].max())))
    dt_min = datetime.fromtimestamp(mktime(gmtime(parsed_item['wall_time'].min())))
    # print(dt_max, dt_min, n_steps, steps_per_value, n_values)
    if not math.isnan(n_steps):
        return str((dt_max - dt_min) / n_steps)
    return False
```

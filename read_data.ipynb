```python
chunks = {'time': 10000}
# temperature_2m = xr.open_mfdataset('./data/all_5.625deg/2m_temperature/2m_temperature_*.nc', chunks=chunks)
# u_wind_10m = xr.open_mfdataset('./data/all_5.625deg/10m_u_component_of_wind/10m_u_component_of_wind_*.nc', chunks=chunks)
# v_wind_10m = xr.open_mfdataset('./data/all_5.625deg/10m_v_component_of_wind/10m_v_component_of_wind_*.nc', chunks=chunks)
geopotential = xr.open_mfdataset('./data/all_5.625deg/geopotential/geopotential_*.nc', chunks=chunks)
# geopotential_500 = xr.open_mfdataset('./data/all_5.625deg/geopotential_500/geopotential_*.nc', chunks=chunks) #single level, vs 13 levels
vorticity_potential = xr.open_mfdataset('./data/all_5.625deg/potential_vorticity/potential_vorticity_*.nc', chunks=chunks)
vorticity = xr.open_mfdataset('./data/all_5.625deg/vorticity/vorticity_*.nc', chunks=chunks)
# relative_humidity = xr.open_mfdataset('./data/all_5.625deg/relative_humidity/relative_humidity_*.nc', chunks=chunks)
specific_humidity = xr.open_mfdataset('./data/all_5.625deg/specific_humidity/specific_humidity_*.nc', chunks=chunks)
temperature = xr.open_mfdataset('./data/all_5.625deg/temperature/temperature_*.nc', chunks=chunks)
# temperature_850 = xr.open_mfdataset('./data/all_5.625deg/temperature_850/temperature_850hPa_*.nc', chunks=chunks) #single level, vs 13 levels
solar_radiation = xr.open_mfdataset('./data/all_5.625deg/toa_incident_solar_radiation/toa_incident_solar_radiation_*.nc', chunks=chunks)
# cloud_cover = xr.open_mfdataset('./data/all_5.625deg/total_cloud_cover/total_cloud_cover_*.nc', chunks=chunks)
precipitation_total = xr.open_mfdataset('./data/all_5.625deg/total_precipitation/total_precipitation_*.nc', chunks=chunks)
u_wind = xr.open_mfdataset('./data/all_5.625deg/u_component_of_wind/u_component_of_wind_*.nc', chunks=chunks)
v_wind = xr.open_mfdataset('./data/all_5.625deg/v_component_of_wind/v_component_of_wind_*.nc', chunks=chunks)

data = xr.merge([
    # temperature_2m,
    # u_wind_10m,
    # v_wind_10m,
    geopotential,
    # geopotential_500,
    vorticity_potential,
    vorticity,
    # relative_humidity,
    specific_humidity,
    temperature,
    # temperature_850,
    solar_radiation,
    # cloud_cover,
    precipitation_total,
    u_wind,
    v_wind
    ])

data
```

```python
levels = [50,  100,  150,  200,  250,  300,  400,  500,  600,  700,  850,  925,
       1000]
var_dict = {
    'z': levels,
    'pv': levels,
    'vo': levels,
    'q': levels,
    't': levels,
    'u': levels,
    'v': levels,
    'tisr': None,
    'tp': None, 
}
generic_level = xr.DataArray([1], coords={'level': [1]}, dims=['level'])
```

```python
data_list = []
for var, levels in var_dict.items():
    print(var, levels)
    try:
        if levels is None:
            data_list.append(data[var].expand_dims({'level': generic_level}, 1))
        else:    
            data_list.append(data[var].sel(level=levels))
    except ValueError:
        print('value error', var)
        data_list.append(data[var].expand_dims({'level': generic_level}, 1))
    except KeyError:
        data_list.append(data[var])
```

```python
data_list
```

```python
_data = xr.concat(data_list, 'level').transpose('time', 'lat', 'lon', 'level')
_data
```

```python
sphere_graph = pickle.load(open('./sphere_graph.pickle', 'rb'))
sphere_graph
```

```python
nodes = [(u, v) for u, v in sphere_graph]
```

```python
nodes[0], _data.sel(lat=nodes[0][0], lon=nodes[0][1], method='nearest').isel(time=np.arange(32))
```

```python
node_attributes = {}
for node in nodes:
    node_attributes[node] = {'data': _data.sel(lat=node[0], lon=node[1], method='nearest')
                             .isel(time=np.arange(1)).load()}
```

```python
numpy_attrs = {}
for node in nodes:
    numpy_attrs[node] = {'x': node_attributes[node]['data'].values.reshape(-1)}
```

```python
numpy_attrs[nodes[0]]['x'].shape
```

```python
with_attrs = sphere_graph.copy()
nx.set_node_attributes(with_attrs, numpy_attrs)
```

```python
from torch_geometric.utils.convert import from_networkx
g = from_networkx(with_attrs, group_edge_attrs=['distance'], group_node_attrs=['x'])
```
